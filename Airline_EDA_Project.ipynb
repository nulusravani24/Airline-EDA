# Airline Passenger Satisfaction – EDA Project

## Dataset Overview
This notebook performs exploratory data analysis on airline passenger satisfaction data.

```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('/mnt/data/airlines_satisfaction_project.CSV-1.xlsx')
df.head()
```

## Overall Satisfaction Distribution

```
df['satisfaction'].value_counts(normalize=True).plot(kind='bar')
plt.ylabel('Proportion')
plt.show()
```

## Average Arrival Delay by Airline

```
df.groupby('Airline')['Arrival Delay in Minutes'].mean().sort_values().plot(kind='bar')
plt.ylabel('Minutes')
plt.show()
```

## Satisfaction by Type of Travel

```
pd.crosstab(df['Type of Travel'], df['satisfaction'], normalize='index')['satisfied'].plot(kind='bar')
plt.ylabel('Proportion')
plt.show()
```

## Conclusion
Business travelers are more satisfied, airlines with lower delays perform better, and punctuality is key to customer satisfaction.
